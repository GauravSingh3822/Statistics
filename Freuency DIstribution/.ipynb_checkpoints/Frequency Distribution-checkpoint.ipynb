```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data=sns.load_dataset("mpg")
```

```python
data
```

```python
data.head()
```

Ungrouoed frequency

```python
data["origin"].value_counts()
```

```python
data["origin"].value_counts().plot(kind="bar")
plt.title("Bar_Graph")
plt.xlabel("Origin")
plt.ylabel("Numbers")

plt.show()
```

```python
data["model_year"].value_counts()

data["model_year"].value_counts().plot(kind="bar")
plt.title("Bar_Graph")
plt.xlabel("Origin")
plt.ylabel("Numbers")

plt.show()
```

```python
pd.crosstab(data["origin"],"Number of cars")
```

```python
pd.crosstab(data["origin"],data["model_year"])
```

Relative Frequency

```python
cars_in_country=pd.crosstab(data["origin"],"Number of cars")
```

```python
cars_in_country["relative frequency"]=cars_in_country["Number of cars"]/sum(cars_in_country["Number of cars"])
```

```python
cars_in_country
```

Percentage Frequency

```python
cars_in_country["Percentage Frequency"]=cars_in_country["relative frequency"]*100
```

```python
cars_in_country
```

```python
data["origin"].value_counts().plot(kind='pie', autopct="%1.f%%")
```
